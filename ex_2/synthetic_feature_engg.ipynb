```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import random
```

```python
np.random.seed(42)
```

```python
values = np.random.randn(200)
```

```python
values
```

```python
dates = pd.date_range(start = '2025/01/01', end = '2025/07/19',freq = 'D')
```

```python
len(dates)
```

```python
dates
```

```python
df = pd.DataFrame(values, index = dates, columns = ['value'])
```

```python
df
```

```python
df['value'].plot(kind = 'hist', bins = 20, title = 'Values')
```

```python
df['value'].plot(kind = 'line', figsize = (8,4), title = 'Values')
```

```python
plt.figure(figsize = (8,4))
plt.plot(df['value'])
plt.xlabel('Date')
plt.ylabel('Freq')
plt.title('Synthetic Time Series Dataset')
plt.show()
```

```python
df['year'] = df.index.year
df['month'] = df.index.month
df['day'] = df.index.day
df['weekday'] = df.index.weekday
df
```

```python
df['lag_01']  = df['value'].shift(1)
```

```python
df
```

```python
df['lag_02']  = df['value'].shift(2)
df
```

```python
df['rolling_mean_5'] = df['value'].rolling(5).mean()
df['rolling_std_5'] = df['value'].rolling(5).std()
df['rolling_min_5'] = df['value'].rolling(5).min()
df['rolling_max_5'] = df['value'].rolling(5).max()
df
```

```python
df['expanding_mean'] = df['value'].expanding().mean()
df['expanding_std'] = df['value'].expanding().std()
df['expanding_min'] = df['value'].expanding().min()
df['expanding_max'] = df['value'].expanding().max()
df
```
