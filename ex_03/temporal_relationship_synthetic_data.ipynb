```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/kaggle/input/min-temp/daily-min-temperatures.csv')
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
Year = df['Date'].dt.year
```

```python
df['Year'] = df['Date'].dt.year
```

```python
df
```

```python
df = df.set_index(df['Year'])
```

```python
df
```

```python
plt.figure(figsize = (8,3))
plt.plot(df['Date'], df['Temp'])
plt.title('Daily Temp bw 1981-1991')
plt.xlabel('Date')
plt.ylabel('Temp in Celsius')
plt.show()
```

```python
plt.figure(figsize = (8,3))
plt.scatter(df['Date'], df['Temp'])
plt.title('Daily Temp bw 1981-1991')
plt.xlabel('Date')
plt.ylabel('Temp in Celsius')
plt.show()
```

```python
plt.figure(figsize = (8,3))
pd.plotting.autocorrelation_plot(df['Temp'])
plt.title('Autocorrelation Plot')
plt.show()
```

```python
plt.figure(figsize = (8,3))
plt.hist(df['Temp'], bins = 10)
plt.title('Histogram for Temperature bw 1981-1991')
plt.xlabel('Temperature in Celsius')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize = (8,3))
sns.kdeplot(df['Temp'], fill = True)
plt.title('Density plot for Temperature bw 1981-1991')
plt.xlabel('Temperature')
plt.show()
```

```python
plt.figure(figsize = (3,2))
sns.boxplot(x = df['Temp'])
plt.title('Box and Whisker plot for Temperature bw 1981-1991')
plt.xlabel('Temperature')
plt.show()
```
