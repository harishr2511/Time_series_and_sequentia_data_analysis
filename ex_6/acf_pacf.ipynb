```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# User Defined Data
drange = pd.date_range(start='1/1/2024', end='12/31/2024', freq='D')
drange
```

```python
mean = (16+40)/2
sd = (40-16)/2
temp_data = np.random.normal(loc=mean, scale=sd-4, size=len(drange))
#temp_data = np.clip(temp_data, 16, 40)
temp_data
```

```python
temp_df = pd.DataFrame({'Date':drange,'Temperature':temp_data})
temp_df
```

```python
temp_df.set_index('Date', inplace=True)
temp_df.head()
```

```python
temp_df.plot(title='Daily Minimum Temperature', figsize=(10,3), legend=None)
plt.xlabel('Date')
plt.ylabel('Temp')
plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# plot ACF and PACF
fig,axes = plt.subplots(2,1,figsize=(10,8))
plot_acf(temp_df['Temperature'],lags=30,ax=axes[0],title='Autocorrelation (ACF)')
plot_pacf(temp_df['Temperature'],lags=30,ax=axes[1],title='Partial Autocorrelation Function (PACF)')
plt.tight_layout()
plt.show()
```
