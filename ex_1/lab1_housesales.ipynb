```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('ma_lga_12345.csv')
```

```python
#first few rows to peek at the data
df.head(10)
```

```python
df.tail()
```

```python
#summary statitsics\
df.describe()
```

```python
df['saledate'] = pd.to_datetime(df['saledate'])
```

```python
df['saledate']
```

```python
df['year'] = df['saledate'].dt.year
df['year']
```

```python
#Filter data for a specific year
filtered_2008 = df[df['year'] == 2008]
filtered_2008
```

```python
#filtering for a particular month
df['month'] = df['saledate'].dt.month
df
```

```python
df['month']
```

```python
#filtering for a particular month
filtered_mar = df[df['month'] == 3]
filtered_mar
```

```python
# 4. Plot average price per year
plt.figure(figsize=(12, 6))
yearly_avg = df.groupby('year')['MA'].mean()
yearly_avg.plot(marker='o')
plt.title('Average Property Price by Year')
plt.xlabel('Year')
plt.ylabel('Average Price')
plt.grid(True)
plt.show()
```

```python
df.head(10)
```

```python
#6. Count of properties sold per year
count_per_yr = df.groupby('year')['type'].count()
count_per_yr
```

```python
#line plot count of properties sold per year
plt.figure(figsize=(12, 6))
count_per_yr.plot()
plt.title('Number of Properties Sold by Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

```python
#bar plot of count per yr
plt.figure(figsize=(12, 6))
yearly_count = df['year'].value_counts().sort_index()
yearly_count.plot(kind='bar')
plt.title('Number of Properties Sold by Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

```python
#Query for a specific date range (e.g., Jan 2010 to Dec 2015)
start_date = '2010-01-01'
end_date = '2015-12-31'
date_range_data = df[(df['saledate'] >= start_date) & (df['saledate'] <= end_date)]
print(f"\nData between {start_date} and {end_date}:")
date_range_data
```

```python
#Calculate the mean price month-wise (use Groupby)
filter_month = df.groupby('month')['MA'].mean()
filter_month
```

```python
#Calculate the mean price month-wise (use Groupby)
filter_month = df.groupby(['year', 'month'])['MA'].mean()
filter_month
```

```python
# Perform a histogram plot
plt.figure(figsize=(12, 6))
plt.hist(df['MA'], bins=50, edgecolor='black')
plt.title('Distribution of Property Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
#Print the property price > 5Lakhs 
filter_price = df[df['MA'] > 500000]
filter_price
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='type', y='MA', data=df)
plt.title('Price Distribution by Property Type')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['saledate'], df['MA'])
plt.title('Property Price Trends Over Time')
plt.xlabel('Sale Date')
plt.ylabel('MA Price')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
